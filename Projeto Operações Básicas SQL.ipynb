Para nossa atividade inicial de SQL trabalharemos com uma base de vendas.
Abaixo temos o ambiente de preparo para trabalharmos com o SQL lite.

```python
import sqlite3 #Importando a biblioteca necessÃ¡ria
import pandas as pd
```

```python
df_vendas = pd.read_csv("TB_VENDAS_TAREFA.csv", delimiter=';')
```

```python
conn = sqlite3.connect(':memory:')
df_vendas.to_sql('tb_vendas', conn, index=False, if_exists='replace')
```

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
query = """
SELECT
*
FROM TB_VENDAS
"""
result_df = run_query(query)
print(result_df)
```

## 3) FaÃ§a uma consulta que retorne uma coluna com os produtos distintos e o valor_unid de cada produto. PorÃ©m apenas para produtos onde o valor_unid Ã© maior ou igual a 50 reais.

<font color='darkblue'>**InterpretaÃ§Ã£o**</font>

<font color='darkblue'>Colunas: PRODUTO, VALOR_UNID</font>

<font color='darkblue'>Filtros: Sim, VALOR_UNID</font>

<font color='darkblue'>Agrupamento:NÃ£o</font>

<font color='darkblue'>OrdenaÃ§Ã£o: NÃ£o</font>

<font color='darkblue'>Distintos: Sim, PRODUTO</font>

<font color='darkblue'>Limite: NÃ£o</font>

<font color='darkblue'>Origem: tb_vendas</font>

```python
query = """
SELECT
DISTINCT PRODUTO,
CAST(VALOR_UNID AS INT) AS VALOR_INT
FROM TB_VENDAS
WHERE VALOR_INT >= 50
"""
result_df = run_query(query)
print(result_df)
```

## [DESAFIO] 4) FaÃ§a uma consulta que retorne o ID das compras e o valor total gasto nessa compra (Valor_unidade X Unidades) das 5 compras com mais valor total gasto.

```python
query = """
SELECT
ID_COMPRA,
SUM(CAST(VALOR_UNID AS INT) * UNIDADES) AS VALOR_TOTAL
FROM TB_VENDAS
GROUP BY ID_COMPRA
ORDER BY VALOR_TOTAL DESC
LIMIT 5
"""
result_df = run_query(query)
print(result_df)
```

## 5) FaÃ§a uma consulta que retorne os produtos e a mÃ©dia do preÃ§o da unidade dos produtos, ordenando do maior para o menor. 

<font color='darkblue'>**InterpretaÃ§Ã£o**</font>

<font color='darkblue'>Colunas: PRODUTO, VALOR_UNID âž” AVG(VALOR_UNID) AS MEDIA_VALOR</font>

<font color='darkblue'>Filtros: NÃ£o</font>

<font color='darkblue'>Agrupamento: Sim, PRODUTO</font>

<font color='darkblue'>OrdenaÃ§Ã£o: Sim, MEDIA_VALOR ASC</font>

<font color='darkblue'>Distintos: NÃ£o</font>

<font color='darkblue'>Limite: NÃ£o</font>

<font color='darkblue'>Origem: tb_vendas</font>

```python
query = """
SELECT
PRODUTO,
AVG(VALOR_UNID) AS MEDIA_VALOR
FROM TB_VENDAS
GROUP BY PRODUTO
ORDER BY MEDIA_VALOR ASC
"""
result_df = run_query(query)
print(result_df)
```

## 6) FaÃ§a uma consulta que retorne o id dos 3 clientes da base de vendas que mais relalizaram compras e a quantidade de compras realizadas.

<font color='darkblue'>**InterpretaÃ§Ã£o**</font>

<font color='darkblue'>Colunas: ID_CLIENTE, ID_COMPRA âž” contar o nÃºmero de compras (COUNT(ID_COMPRA)) AS QTD_COMPRAS</font>

<font color='darkblue'>Filtros: NÃ£o</font>

<font color='darkblue'>Agrupamento: Sim âž” sem GROUP BY, o SQL calcula apenas um Ãºnico resultado agregado da tabela toda.</font>

<font color='darkblue'>OrdenaÃ§Ã£o: Sim, QTD_COMPRAS âž” decrescente</font>

<font color='darkblue'>Distintos: NÃ£o</font>

<font color='darkblue'>Limite: Sim, 3</font>

<font color='darkblue'>Origem: tb_vendas</font>

```python
#contar o nÃºmero de compras (COUNT(ID_COMPRA))
#sem GROUP BY, o SQL calcula apenas um Ãºnico resultado agregado da tabela toda.
query = """
SELECT
ID_CLIENTE,
COUNT(ID_COMPRA) AS QTD_COMPRAS
FROM TB_VENDAS
GROUP BY ID_CLIENTE  
ORDER BY QTD_COMPRAS DESC
LIMIT 3
"""
result_df = run_query(query)
print(result_df)
```
