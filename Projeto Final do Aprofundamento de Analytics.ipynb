**Tabelas Disponibilizadas:**

**Tabela de Transações:** Contém os registros de transações realizadas pelos clientes, incluindo detalhes como ID da transação, valor e outros.


**Tabela de Dados Pessoais:** Contém as informações pessoais dos clientes, como ID do cliente, nome, genero, cidade, etc.

**Chave de Ligação:** As tabelas se relacionam através da coluna ID_CLIENT, que é a chave identificadora dos clientes.

# Etapas do Projeto:

Abaixo temos a configuração do ambiente SQL:

```python
import sqlite3
import pandas as pd
```

```python
df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
#COLUNA TB_CLIENTES
conn = sqlite3.connect('projeto.db')
query = """
SELECT
*
FROM TB_CLIENTES
"""

df = pd.read_sql(query, conn)
df
```

```python
#COLUNA TB_TRANSACOES
conn = sqlite3.connect('projeto.db')
query = """
SELECT
*
FROM TB_TRANSACOES
"""
df = pd.read_sql(query, conn)
df
```

```python
# COLUNAS TB_CLIENTES + TB_TRANSACOES
query_completa = """
SELECT DISTINCT
A.*,
B.*
FROM TB_CLIENTES AS A
FULL JOIN TB_TRANSACOES AS B
ON A.id_client = B.id_client
"""
df_completa = run_query(query_completa)
print(df_completa)
```

```python
# COLUNAS TB_CLIENTES + TB_TRANSACOES COM VALORES FALTANTES
conn = sqlite3.connect('projeto.db')
query_faltantes = """
SELECT DISTINCT
*
FROM TB_CLIENTES AS A
FULL OUTER JOIN TB_TRANSACOES AS B
ON A.id_client = B.id_client
WHERE A.id_client IS NULL OR B.id_client IS NULL
"""
df_faltantes = run_query(query_faltantes)
print(df_faltantes)
```

 * <font color='darkblue'> 'A.id_client = valores preenchidos' + 'B.id_client = nulos' → clientes sem transações.</font>
 * <font color='darkblue'>'A.id_client = nulo' + 'B.id_client = preenchido' → transações sem cliente cadastrado.</font>

```python
# COLUNAS TB_CLIENTES + TB_TRANSACOES SEM VALORES FALTANTES
conn = sqlite3.connect('projeto.db')
query_consolidada = """
SELECT DISTINCT
*
FROM TB_CLIENTES AS A
INNER JOIN TB_TRANSACOES AS B
ON A.id_client = B.id_client
WHERE A.id_client IS NOT NULL AND B.id_client IS NOT NULL
"""

df_consolidada = run_query(query_consolidada)
print(df_consolidada)
```

Justifique a escolha do JOIN:

* <font color='darkblue'> Foi utilizado o 'FULL JOIN', pois o objetivo da tabela é consolidar todas as informações, tendo dados faltantes ou não.</font>


* <font color='darkblue'> Para criar uma tabela em que há 'id_cliente' faltando dados, foi usado o 'FULL OUTER JOIN' em ambas as colunas com o filtro 'WHERE A.id_client IS NOT NULL AND B.id_client IS NOT NULL' para conseguir visualizar somente os clientes que estão com valores faltantes.</font>


* <font color='darkblue'> Em uma nova tabela, para visualizar somente os 'id_cliente' com dados, foi usado o 'INNER JOIN' com um filtro diferente 'WHERE A.id_client IS NOT NULL AND B.id_client IS NOT NULL'. Assim a tabela só traz os 'id_cliente' completos.</font>

Exportando o arquivo como CSV:

```python
df_consolidada = pd.read_sql(query_consolidada, conn)

df_consolidada.to_csv('dados_consolidados.csv', index=False, encoding='utf-8-sig')

print("Arquivo 'dados_consolidados.csv' exportado com sucesso!")
```

```python
df_faltantes = pd.read_sql(query_faltantes, conn)

df_faltantes.to_csv('dados_faltantes.csv', index=False, encoding='utf-8-sig')

print("Arquivo 'dados_faltantes.csv' exportado com sucesso!")
```

```python
df_completa = pd.read_sql(query_completa, conn)

df_completa.to_csv('dados_completos.csv', index=False, encoding='utf-8-sig')

print("Arquivo 'dados_completos.csv' exportado com sucesso!")
```
